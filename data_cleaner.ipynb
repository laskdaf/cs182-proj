```python
directory = './train_sentiment'
destination = 'json_with_labels'

try:  
    os.mkdir(destination)
except OSError:  
    print ("Creation of the directory %s failed")
else:  
    print ("Successfully created the directory %s ")

for filename in os.listdir(directory):
    if filename.endswith(".json"):
        with open('./train_sentiment/' + filename) as json_file:
            print(filename)
            data = json.load(json_file)
            
            # keep everything english
            if (data["language"] != "en"):
                continue
                
            pet_hash = filename[:-5]
            query = pet_data[pet_data['PetID']==pet_hash]
            print(pd.to_numeric(query["Type"]))
            print(type(pd.to_numeric(query["Type"])))
            
            age = query['Age'].item()
            breed1 = query['Breed1'].item()
            breed2 = query['Breed2'].item()
            gender = query['Gender'].item()
            color1 = query['Color1'].item()
            color2 = query['Color2'].item()
            color3 = query['Color3'].item()
            maturitySize = query['MaturitySize'].item()
            furLength = query['FurLength'].item()
            vaccinated = query['Vaccinated'].item()
            dewormed = query['Dewormed'].item()
            sterilized = query['Sterilized'].item()
            quantity = query['Quantity'].item()
            fee = query['Fee'].item()
            photoAmt = query['PhotoAmt'].item()
            label = query['AdoptionSpeed'].item()

            data['Age'] = age
            data['Breed1'] = breed1
            data['Breed2'] = breed2
            data['Gender'] = gender
            data['Color1'] = color1
            data['Color2'] = color2
            data['Color3'] = color3
            data['MaturitySize'] = maturitySize
            data['FurLength'] = furLength
            data['Vaccinated'] = vaccinated           
            data['Dewormed'] = dewormed
            data['Sterilized'] = sterilized
            data['Quantity'] = quantity           
            data['Fee'] = fee
            data['PhotoAmt'] = photoAmt
            data['label'] = label
    
            
            with open('./' + destination + "/with_label_" + filename, 'w') as outfile:
                json.dump(data, outfile, indent=4)
```
